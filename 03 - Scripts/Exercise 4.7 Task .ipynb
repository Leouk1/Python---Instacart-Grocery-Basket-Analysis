# If-Statements with User-Defined Functions

```python
## Create subset of first 1,000,000 rows
df = ords_prods_merge[:1000000]
```

```python
## code for price_labels
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
## Applying to data frame 
df['price_range'] = df.apply(price_label, axis=1)
```

```python
## count values in price range column
df['price_range'].value_counts(dropna=False)
```

```python
## check most expensive product
df['prices'].max()
```

# 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”


```python
## Find highest frequency hour of the order_hour_day
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

## I have identified the frequency of each hour. I have split this up into the 3 lables requested by client:
most orders: hours 9-16
average orders: 17-20
fewest orders: all other orders

why?
I grouped the hours into “Most,” “Average,” and “Fewest orders” based on the frequency of orders, with 9–16 showing the highest volumes (daytime peak), 17–20 showing moderate activity (evening), and 21-8 showing the lowest volumes (overnight and early morning).

The values for each label can be modified upon request.

```python
## Making lables and assigning values for new column busiest_period_of_day
result_4 = []

for value in ords_prods_merge["order_hour_of_day"]:
    if 9 <= value <= 16:          # hours 9–16 → Most orders
        result_4.append("Most orders")
    elif 17 <= value <= 20:       # hours 17–20 → Average orders
        result_4.append("Average orders")
    else:                         # all other hours → Fewest orders
        result_4.append("Fewest orders")
```

```python
## Making new column busiest_period_of_day with above lables and values
ords_prods_merge["busiest_period_of_day"] = result_4
```
