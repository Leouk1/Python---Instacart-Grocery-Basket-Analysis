# 6.The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
1. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
2. If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
## Calculate each user’s average price
ords_prods_merge['average_price'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
## Create the spending_flag column
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
## Check results
ords_prods_merge['spending_flag'].value_counts()
```

```python
## Check new columns
ords_prods_merge.head(4)
```

```python
## Drop average_price as client does not need it
ords_prods_merge.drop(columns=['average_price'], inplace=True)
```

# 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
   
1. If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
2. If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
3. If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
pd.options.display.max_columns=None
```

```python
## Calculate each users median number of days between orders
ords_prods_merge['median_days'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform('median')
```

```python
## Create the order frequency flag
ords_prods_merge.loc[ords_prods_merge['median_days'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days'] > 10) & (ords_prods_merge['median_days'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
## Check results
ords_prods_merge['order_frequency_flag'].value_counts()
```

```python
ords_prods_merge.head(3)
```

```python
## Dropping median days column as instructions dont specify it is needed.
ords_prods_merge.drop(columns=['median_days'], inplace=True)
```
