# 3.You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.
- In a markdown cell beneath your histogram, describe what the histogram demonstrates.

```python
## Create Histogram and Adding labels
Hist = ords_prods_cust['order_hour_of_day'].plot.hist(bins = 24)
plt.title('Distribution of Orders by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows how customer orders are distributed throughout the day.
Most orders are placed between 9 a.m. and 4 p.m., with the highest order frequency around 10 a.m. to 3 p.m.
There’s a steep drop in orders after 8 p.m., and very few orders occur during the night or early morning hours.
This suggests that Instacart customers primarily place orders during normal daytime shopping periods.

# 4.The marketing team is curious about the distribution of orders among customers in terms of loyalty. Create a bar chart from the “loyalty_flag” column.

```python
## Creating a bar chart for customer loyalty
Bar = ords_prods_cust['loyalty_flag'].value_counts().plot.bar()

plt.title('Distribution of Customers by Loyalty Status')
plt.xlabel('Loyalty Category')
plt.ylabel('Number of Customers')
plt.show()
```

The bar chart shows that regular customers make up the largest share of Instacart users, followed by loyal customers, and then new customers.
This suggests that most of Instacart’s users are returning shoppers who make repeat purchases, while the platform continues to attract new users at a smaller rate.
Overall, the company appears to have a strong base of consistent customers, indicating good user retention.

First, determine whether there’s a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents:
- In a markdown cell beneath your line chart, describe what the line chart shows.

```python
customers = pd.read_pickle(os.path.join(path,'02 - data','prepared data','customers_cleaned.pkl'))
```

```python
## Create line chart showing relationship between age and number of dependents
line_2 = sns.lineplot(data=customers, x='age', y='n_dependants')

plt.title('Relationship Between Age and Number of Dependents')
plt.xlabel('Age')
plt.ylabel('Number of Dependents')
plt.show()
```

This suggests that customers of different ages tend to have similar family sizes on average, although younger and older customers may have slightly fewer dependents.

# 7.You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.
- In a markdown cell beneath your scatterplot, describe what the scatterplot shows.


```python
## Create scatterplot showing relationship between age and income
scatter = sns.scatterplot(data=customers, x='age', y='income')

plt.title('Relationship Between Age and Income')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.show()
```

The scatterplot shows the relationship between customer age and income.
There isn’t a strong visible correlation, but there is a slight upward trend, with older customers generally earning more than younger ones.
Most incomes fall between $50,000 and $150,000, while higher-income outliers are more common among customers aged 40 and above.
This suggests that spending power tends to increase with age, although income levels vary widely across all groups.
